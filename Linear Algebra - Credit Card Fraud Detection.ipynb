```python
X=df.groupby('Class')
X0=X.get_group(0)
not_fraud = X0[:598]
Y=X.get_group(1)
fraud=Y[:2]
credit=pd.concat([fraud,not_fraud])
```

```python
credit.head()
```

```python
detect=np.zeros((600, 599))
```

```python
for i in range(600):
    count=0
    length=np.array([credit.iloc[i]])
#print(length)
    for ele in length**2:
        count=count+ele
    a=sum(count)
    l_vector1=a**(0.5)
    k=0
    for j in range(600):
        
        if i!=j:
            count1=0
            length1=np.array([credit.iloc[j]])
            for elem in length1**2:
                count1=count1+elem
            b=sum(count1)
            l_vector2=b**(0.5)
            detect[i][k]=np.dot(credit.iloc[i],credit.iloc[j])/(l_vector1*l_vector2)
            k=k+1

detect
#print(l_vector1)       
```

```python
b=np.argsort(detect,axis=1)


for i in range(600):
    
    for j in range(10):
        k=b[i][j]
        if i==k or i<k:
            print('class of vector {} is {}'.format(k+1,int(credit.iloc[k+1].Class)))
            print('similarity for vector {} and {} is {} :'.format(i,k+1,detect[i][k]))
            print('\n')    

        else:
            print('class of vector {} is {}'.format(k,int(credit.iloc[k].Class)))
            print('similarity for vector {} and {} is {} :'.format(i,k,detect[i][k]))
            print('\n')  
            
    print("********************************************")
            
        
        
```
